# Exploratory Data Analysis 

# Gender

```python
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='Gender', y='Amount', data =sales_gen )
```

From above graph we can see that most of the buyers are Females

# AGE

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group', data=df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Age group 
```

```python
sales_gen=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data =sales_gen )
```

From above graph we can see that most of the buyers are from Age Group 26-35

# State

```python
df.columns
```

```python
sns.set(rc={"figure.figsize":(25, 10)})
ax=sns.countplot(x='State', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize":(25, 10)})
sales_gen=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='State', y='Amount', data =sales_gen )
```

From above graph we can see that most of the orders are from UttarPradesh, Maharashtra and Karnataka 

# Marital_Status

```python
ax=sns.countplot(x='Marital_Status', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize":(6, 5)})
sales_gen=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='Marital_Status', y='Amount', data =sales_gen, hue='Gender')
```

From above graph we can see that most of the buyers are Married(Female)

# Occupation

```python
sns.set(rc={"figure.figsize":(20, 10)})
ax=sns.countplot(x='Occupation', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize":(25, 10)})
sales_gen=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='Occupation', y='Amount', data =sales_gen )
```

From above graph we can see that most of the buyers are from IT sector, Healthcare,Aviation

```python
sns.set(rc={"figure.figsize":(25, 10)})
ax=sns.countplot(x='Product_Category', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize":(25, 10)})
sales_gen=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='Product_Category', y='Amount', data =sales_gen )
```

From above graph we can see that most the sold products are Food,Clothing and Electronics

```python
sns.set(rc={"figure.figsize":(25, 10)})
sales_gen=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='Product_ID', y='Orders', data =sales_gen )
```

# Conclusion

From all the points we can conclude that Married women from Age Group 26-35 years from UP , Maharashtra, Karnataka working in IT sector and Healthcare are like to buy products like: Food,Clothing
